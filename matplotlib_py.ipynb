```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.
Ans:
    Matplotlib is a popular data visualization library in Python. It provides a wide range of tools and functions for creating 
    various types of plots, charts, and graphs. Matplotlib is widely used for scientific computing, exploratory data analysis,
    and presenting data visually.

Here are five plots that can be created using the Pyplot module of Matplotlib:

Line Plot: Line plots are useful for visualizing trends and relationships between two variables. They are created using the plot
    function and are commonly used to represent time series data.

Bar Plot: Bar plots display categorical data as rectangular bars, where the length of each bar represents the value of a speci-
    fic category. They are created using the bar or barh functions and are often used to compare different categories.

Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables. Each data point is repre-
    sented as a dot on the plot, with the x and y coordinates determined by the variables being compared. Scatter plots are cr-
    eated using the scatter function.

Histogram: Histograms are used to represent the distribution of a single variable. They divide the data into bins and display 
    the frequency or count of data points falling into each bin. Histograms can be created using the hist function.

Pie Chart: Pie charts are circular charts divided into sectors, where each sector represents a proportion of a whole. They are
    useful for representing categorical data as proportions or percentages. Pie charts can be created using the pie function.

These are just a few examples of the types of plots that can be created using Matplotlib's Pyplot module. Matplotlib offers
many more plot types and customization options, allowing you to create a wide variety of visualizations to suit your needs.
```

```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
 Ans:   
A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional coordinate 
system. It is used to visualize the relationship between two continuous variables, where one variable is represented on the 
x-axis and the other on the y-axis. Scatter plots are useful for identifying patterns, correlations, and clusters in the data.

Now, let's use the given code to generate data for x and y and then plot a scatter plot using Matplotlib:

python code"""
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
 Ans:
The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a 
grid-like structure, specifying the number of rows, columns, and the position of each plot within the grid. This is useful when
you want to compare multiple plots side by side or visualize different aspects of the data together.

Now, let's use the subplot() function to draw four line plots with the given data:

python code"""
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Data for line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(12, 8))  # Optional: Set figure size

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='o', label='Line 1')
plt.title('Line Plot 1')
plt.xlabel('X')
plt.ylabel('Y')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, marker='o', label='Line 2')
plt.title('Line Plot 2')
plt.xlabel('X')
plt.ylabel('Y')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, marker='o', label='Line 3')
plt.title('Line Plot 3')
plt.xlabel('X')
plt.ylabel('Y')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, marker='o', label='Line 4')
plt.title('Line Plot 4')
plt.xlabel('X')
plt.ylabel('Y')

# Adjust layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()

```

```python
"""Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
ChatGPT
A bar plot, also known as a bar chart, is a data visualization that uses rectangular bars to represent categorical data. Each
bar's length or height is proportional to the value it represents, and the bars are spaced equally to represent different 
categories. Bar plots are commonly used to compare and visualize the relative values of different categories in a dataset.

Bar plots are useful for the following reasons:

Comparison: Bar plots allow easy visual comparison of the magnitudes of different categories, making it straightforward to 
    identify which category has the highest or lowest value.

Categorical Data: Bar plots are particularly effective for displaying categorical data, where each bar represents a distinct 
    category or group.

Clear Representation: The visual simplicity of bar plots makes it easy to communicate information to a broad audience, making 
    them widely used in various fields for data presentation.

Now, let's create a bar plot and a horizontal bar plot using the given data:

python  code"""
import numpy as np
import matplotlib.pyplot as plt

# Given data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.title('Company Profits (Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit (Millions)')
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.title('Company Profits (Horizontal Bar Plot)')
plt.xlabel('Profit (Millions)')
plt.ylabel('Company')
plt.show()
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
Ans:
    
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of data. It displays a summary of the data's five-number summary: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. The box in the middle represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). The whiskers extend from the box to the minimum and maximum values within a certain range (commonly 1.5 times the IQR).

Box plots are useful for the following purposes:

Outlier Detection: Box plots can effectively identify potential outliers in the data, which are values that significantly differ from the rest of the data.

Comparison: Box plots enable easy comparison of data distributions between different groups or datasets.

Summary Visualization: Box plots provide a concise summary of the data distribution, including measures of central tendency (median) and dispersion (IQR).

Now, let's create a box plot using the given data:

python code"""
import numpy as np
import matplotlib.pyplot as plt

# Given data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data into a list for box plot
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```
